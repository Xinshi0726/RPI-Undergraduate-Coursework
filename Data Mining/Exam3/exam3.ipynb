#### Q1. K-Means: (Total 15 points)  For this question, we will use the first attribute as the true class for evaluation, and the remaining 13 attributes as the actual attributes for use in K-means clustering. Use the minmax scaled data. Answer the following:

##### a) (10 points) Code and run K-Means with $K=3$ using convergence threshold $\epsilon=0.001$. Report the i) final means, ii) the number of points in each cluster, and iii) the final SSE value. Set random seed as 42, and find the initial clusters using the "farthest means" approach, i.e., after the first mean selected at random as one of the data points, the remaining two means should be points that are the farthest from already selected means.

##### b) (5 points) What is the silhouette coefficient for point $x_0$, and the silhouette coefficent for the **ground truth** clustering. Use Euclidean distance for the proximity matrix.

```python
#Part a
def find_farthest_means(data, k):
    dist_to_mean1 = []
    dist_to_mean2 = []
    mean1_idx = np.random.randint(len(data))
    mean1 = data[mean1_idx]
    dist1 = 0
    mean2_idx = 0 
    for idx,i in enumerate(data):
        if (np.linalg.norm(mean1-i)>dist1):
            dist_to_mean1.append(dist1)
            dist1 = np.linalg.norm(mean1-i)
            mean2_idx = idx
    mean2 = data[mean2_idx]
    for idx,i in enumerate(data):
        dist_to_mean2.append(np.linalg.norm(mean2-i))
    dist3 = 0
    mean3_idx = 0
    for i in range(len(dist_to_mean1)):
        if (dist_to_mean1[i]+dist_to_mean2[i])>dist3:
            dist3 = dist_to_mean1[i]+dist_to_mean2[i]
            mean3_idx = i
    mean3 = data[mean3_idx]
    return [mean1,mean2,mean3]
def kmeans(data, k):
    mean_centers = find_farthest_means(data,k)
    while (True):
        distances = np.array([[np.linalg.norm(data[i] - mean_centers[j]) for j in range(k)] for i in range(data.shape[0])])
        clusters = np.argmin(distances, axis=1)
        new_mean_centers = np.array([data[clusters == i].mean(axis=0) for i in range(k)])
        if np.linalg.norm(mean_centers-new_mean_centers)<=0.001:
            return clusters,mean_centers
        mean_centers = np.copy(new_mean_centers)
def sse(data, mean_centers, clusters):
    distances = np.array([np.linalg.norm(data[i] - mean_centers[clusters[i]]) for i in range(data.shape[0])])
    return np.sum(distances ** 2)
```

```python
clusters,centers = kmeans(X,3)
print("The final means are:")
print(centers)
print(f"The cluster distribution is: {[np.sum(clusters == i) for i in range(3)]}")
print(f"The SSE value is: {sse(X,centers,clusters)}")
```

```python
#Part b
def silhouette_coefficient(data, mean_centers, clusters):
    distances = np.array([[np.linalg.norm(data[i] - mean_centers[j]) for j in range(mean_centers.shape[0])] for i in range(data.shape[0])])
    a = np.array([distances[i][clusters[i]].mean() for i in range(data.shape[0])])
    b = np.array([np.min([distances[i][j] for j in range(mean_centers.shape[0]) if j != clusters[i]]) for i in range(data.shape[0])])
    silhouette_coefficients = (b - a) / np.maximum(a, b)
    return silhouette_coefficients
```

#### Q4. Gradient vs. Mean-Shift: (Total 15 points) Given the 1st point (at index 0), $x_0$ in the dataset (use the minmax scaled data) answer the following questions:

##### a) (CSCI4390: 15 points, CSCI6390: 10 points) Compute the density attractor for $x_0$ via mean-shift update, Use $h=0.5, \epsilon=0.001$. How many iterations does it take to converge. Print the attractor and its density.

##### b) (**CSCI6390 Only**: 5 points) Compute the density attractor for $x_0$ via gradient update, Use $h=0.5,  \epsilon=0.001, \eta=1$. How many iterations does it take to converge. Print the attractor and its density. Which of the two do you think is the better method, and why?

```python
import math
def Gaussian_Kernel(x_t,x_i,h):
    return (1/((2*math.pi)**(64/2)))*np.exp(-np.sum((x_i-x_t)*(x_i-x_t),axis = 1, keepdims=True)/(2*h**2))
def findAttractor(x,D,h,eps):
    x_prev = np.copy(x)
    x_next = np.copy(x_prev)
    t = 0
    while t == 0 or np.linalg.norm(x_next-x_prev)>=eps:
        x_prev = np.copy(x_next)
        k_i = Gaussian_Kernel(x_next,D,h)
        numerator = np.sum((k_i*D),axis = 0)
        denominator= np.sum(k_i)
        x_next = numerator/denominator
        t+=1
    return x_next,t
def f_hat(x_star,n,h,d,D):
    partial = np.sum(Gaussian_Kernel(x_star,D,h))*1/(n*h**d)
    return partial
```

```python
x_star,t = findAttractor(X[0,:],X,0.5,0.001)
density = f_hat(x_star,X.shape[0],0.5,13,X)
print(f"It takes {t} iterations to converge")
print(f"The attractor is:")
print(x_star)
print(f"The density is:")
print(density)
```

However, after convergence, report only the attractor with the largest number of points. List the attractor index, and the list of points attracted to it.

```python
def gamma(M,r):
    new_M = np.zeros(M.shape)
    for i in range(M.shape[0]):
        denominator = np.sum(M[i])
        for j in range(M.shape[1]):
            new_M[i][j] = (M[i][j]**r)/denominator
    return new_M
h = 0.1
norm = squareform(pdist(X, 'euclidean'))
A = np.exp(-(norm ** 2)/(2*h))
degree_matrix_inv = (1/np.sum(A, axis=0)) * np.identity(X.shape[0])
M_t = np.dot(degree_matrix_inv,A)
t = 0
M_t_1 = np.copy(M_t)
while (t == 0 or np.linalg.norm(M_t_1 - M_t)>0.001):
    M_t = np.copy(M_t_1)
    t += 1
    M_t_1 = np.dot(M_t.T,M_t)
    M_t_1 = gamma(M_t_1,3)
```
